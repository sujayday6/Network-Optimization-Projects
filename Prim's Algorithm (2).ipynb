```python
# Prim's Algorithm to get Minimum Spanning Tree
'''
# Instructions To Run the Code
1. Change the csv file name 
2. Now run the program, you will get MST and running time in the output 
'''
from collections import defaultdict
import pandas as pd

# Here change the csvfile name
csv='100Nodes.csv'
 
gdf=pd.read_csv(csv)   # Converting csvfile into the Pandas Dataframe
nvertices=100  # Finding the number of vertices/nodes in the given graph

# For the Prim's Algorithm we need to convert the graph data into Adj Matrix such that it will have value 
#if the there's an edge between the column number vertex and row number vertex
# Creating the empty list called AdjMatrix
adjMatrix = []
# Now making the matrix of size n*n (n= number of vertices in graph)
for i in range(0, nvertices):
    adjMatrix.append([])
    for j in range(0, nvertices):
        adjMatrix[i].append(0)
# Giving values for the element if there's an edge exists
for i in range(0, len(gdf)):
    a= gdf.iloc[i,0]
    b= gdf.iloc[i,1]
    adjMatrix[a][b] = gdf.iloc[i,2]
    adjMatrix[b][a] = gdf.iloc[i,2]
    

class Graph():

    def __init__(self,vertices):
        self.V= vertices # Number of vertices  
    
    # Function to get the MST by using Prim's Algorithm
    def PrimMST(self):
        # Randomly choosing any vertex for starting the graph
        vertex = 0
  
        MST = [] # list MST to store the resultant MSt
        pedges = [] # list for the adding the possible edges from the particular vertex
        visited = [] # List to store visited vertices
        minEdge = [None,None,float('inf')]
  
      # Stop when until you get Spanning Tree having the number of edges |v|-1
        while len(MST) != self.V -1:
    
        # Adding the vertex to visited list 
            visited.append(vertex)
    
        # Adding the all possible edges into the list edges 
        # We will add the edge if the weight value is not equal to zero
            for i in range(0, self.V):
                if adjMatrix[vertex][i] != 0:
                    pedges.append([vertex,i,adjMatrix[vertex][i]])
        
        # We will add the edge with minimum weight and which one of vertex is not visited yet!
            for e in range(0, len(pedges)):
                if pedges[e][2] < minEdge[2] and pedges[e][1] not in visited:
                    minEdge = pedges[e]
        
        # Removing edge with minimum weight from list of edges
            pedges.remove(minEdge)

        # And adding that edge to the MST 
            MST.append(minEdge)
      
            vertex = minEdge[1] # Resetting the vertex value   
            minEdge = [None,None,float('inf')] # Making the minEdge value to none
        
        
        # To calulate the total weight for the resultant MST
        total_cost=0
        print ("By Prim's algorithm we get following MST")
        for i in range(0,len(MST)):
            print ("Edge %d - %d " % (MST[i][0],MST[i][1]))
            total_cost= total_cost + MST[i][2] # Updating the total weight value for each edge in the resultant MST
        
        
        print("Minimum Total Weight = %d" %total_cost)


Graph(nvertices).PrimMST()
```
